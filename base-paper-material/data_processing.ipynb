```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

dire = ["trained_pereira",
       "trained_fedorenko",
       "trained_blank",
       "untrained_pereira",
       "untrained_fedorenko",
       "untrained_blank"]


for name in dire:
    string = name + "/model_info.csv"
    df = pd.read_csv(string)
    print(f'This is for {string}:')
    print("layers vs scores:", df['layers'].corr(df['max scores'], method='pearson'))
    print("neurons vs scores:", df['neurons'].corr(df['max scores'], method='pearson'))
    print("neurons*layer vs scores:", df['neurons*layer'].corr(df['max scores'], method='pearson'))
    print("mean of the scores vs scores:", df['mean of the scores'].corr(df['max scores'], method='pearson'))
    print("median of the scores vs scores:", df['median of the scores'].corr(df['max scores'], method='pearson'))
    print("variance of the scores vs scores:", df['variance of the scores'].corr(df['max scores'], method='pearson'))


    
    # --- 1. Individual scatter plots ---
    features = [
        'layers',
        'neurons',
        'neurons*layer',
        'mean of the scores',
        'median of the scores',
        'variance of the scores'
    ]


    
    for feature in features:
        corr_value = df[feature].corr(df['max scores'], method='pearson')
    
        plt.figure(figsize=(6, 4))
        
        # Scatter + regression line
        sns.regplot(
            x=df[feature], 
            y=df['max scores'], 
            scatter_kws={'alpha': 0.7},
            line_kws={'color': 'red', 'lw': 2}
        )
        
        plt.title(f"{feature} vs max scores\nPearson r = {corr_value:.3f}")
        plt.xlabel(feature)
        plt.ylabel('max scores')
        plt.grid(True, linestyle='--', alpha=0.5)
        
        filename = f"{feature.replace('*', 'x').replace(' ', '_')}_vs_max_scores.png"
        plt.savefig(os.path.join(name, filename), dpi=300, bbox_inches='tight')
        plt.close()
    
    print("Scatter plots saved.")
    
    # --- 2. Correlation heatmap ---
    corr = df[features + ['max scores']].corr()
    
    plt.figure(figsize=(12, 10))
    sns.heatmap(
        corr,
        annot=True,
        cmap='coolwarm',
        annot_kws={"size": 12},
        fmt=".2f"
    )
    
    plt.title("Correlation Matrix", fontsize=16)
    plt.xticks(rotation=45, ha="right", fontsize=12)
    plt.yticks(rotation=0, fontsize=12)
    
    plt.tight_layout()
    plt.savefig(os.path.join(name, "correlation_matrix.png"), dpi=300)
    plt.show()
    plt.close()
    
    print("Correlation matrix saved.")
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os

save_dir = "/Users/nezhadinho/Desktop/Saarland/S03/NNBC/Seminar report/trained_pereira"
os.makedirs(save_dir, exist_ok=True)

features = ['layers', 'neurons', 'neurons*layer', 'mean of the scores', 'median of the scores', 'variance of the scores']
target = 'max scores'

for feature in features:
    # Coerce to numeric and align valid pairs only
    x_raw = pd.to_numeric(df[feature], errors='coerce')
    y_raw = pd.to_numeric(df[target], errors='coerce')
    mask = np.isfinite(x_raw) & np.isfinite(y_raw)
    x = x_raw[mask].to_numpy()
    y = y_raw[mask].to_numpy()

    # Correlation (pairwise over the cleaned data)
    corr = np.corrcoef(x, y)[0, 1] if x.size > 1 else np.nan

    plt.figure(figsize=(6, 5))
    plt.scatter(x, y, alpha=0.7)

    # Best-fit line only if we have at least 2 points and x varies
    if x.size >= 2 and np.std(x) > 0:
        m, b = np.polyfit(x, y, 1)
        order = np.argsort(x)
        plt.plot(x[order], m * x[order] + b, linestyle='--', label=f'corr = {corr:.3f}')
        plt.legend()
    else:
        plt.text(0.05, 0.95, 'Not enough valid/variable data for fit',
                 transform=plt.gca().transAxes, va='top')

    plt.xlabel(feature)
    plt.ylabel('scores')
    plt.title(f'{feature} vs scores')
    plt.grid(True)
    plt.tight_layout()

    filename = f"{feature.replace('*', 'x').replace(' ', '_')}_vs_{target.replace(' ', '_')}.png"
    plt.savefig(os.path.join(save_dir, filename))
    plt.show()
```

```python
import pandas as pd

directory = ["trained_pereira",
       "trained_fedorenko",
       "trained_blank",
       "untrained_pereira",
       "untrained_fedorenko",
       "untrained_blank"]


for name in directory:
    string = name + "/model_info.csv"
    # Load your dataset
    df = pd.read_csv(string)

    # Group by layers and neurons, collect model names into a list
    grouped_df = (
        df.groupby(["layers", "neurons"])
          .agg(models=("model", lambda x: x.tolist()))  # Changed list(x) to x.tolist() to avoid list name conflict
          .reset_index()
          .sort_values(["layers", "neurons"], kind="mergesort")
    )
    
    # Save to CSV
    stringg = name + "/grouped_models.csv"
    grouped_df.to_csv(stringg, index=False)
    
    print(f"Grouped models saved to {stringg}")
```

```python
def split_model(string):
    list = []
    string = str(string.replace("[", "").replace("]", "").replace("'", "").replace(" ", ""))
    list = string.split(",")
    return list
    

directory = ["trained_pereira",
       "trained_fedorenko",
       "trained_blank",
       "untrained_pereira",
       "untrained_fedorenko",
       "untrained_blank"]
for name in directory:
    # model_info
    string = name + "/model_info.csv"
    df_model_info = pd.read_csv(string)

    # grouped_models
    string = name + "/grouped_models.csv"
    df_grouped_models = pd.read_csv(string)


for item in df_grouped_models['models']:
    new_df = df_model_info[df_model_info["model"].isin(split_model(item))]
    print(f'This are the models: {split_model(item)}')
    print("layers vs scores:", new_df['layers'].corr(new_df['max scores'], method='pearson'))
    print("neurons vs scores:", new_df['neurons'].corr(new_df['max scores'], method='pearson'))
    print("neurons*layer vs scores:", new_df['neurons*layer'].corr(new_df['max scores'], method='pearson'))
    print("mean of the scores vs scores:", new_df['mean of the scores'].corr(new_df['max scores'], method='pearson'))
    print("median of the scores vs scores:", new_df['median of the scores'].corr(new_df['max scores'], method='pearson'))
    print("variance of the scores vs scores:", new_df['variance of the scores'].corr(new_df['max scores'], method='pearson'))
```

```python
# 1) Keep only model + max score
t = data_trained[['model', 'max scores']].copy()
u = data_untrained[['model', 'max scores']].copy()

# 2) Normalize model names so they match
t['model_base'] = t['model']  # trained names already base
u['model_base'] = u['model'].str.replace(r'-untrained$', '', regex=True)

# 3) Merge trained â†” untrained by base name
pairs = (t.rename(columns={'max scores':'max_trained'})
           .merge(u.rename(columns={'max scores':'max_untrained'}),
                  on='model_base', how='inner', suffixes=('_trained','_untrained')))

# Optional: check pairs
# print(pairs[['model_base','max_trained','max_untrained']])

# 4) Correlations
pearson  = pairs['max_trained'].corr(pairs['max_untrained'], method='pearson')
spearman = pairs['max_trained'].corr(pairs['max_untrained'], method='spearman')
kendall  = pairs['max_trained'].corr(pairs['max_untrained'], method='kendall')

print({'pearson': pearson, 'spearman': spearman, 'kendall': kendall})
```
